```python
import pandas as pd
import pymongo
from pymongo import MongoClient
```

```python
#Importing data

prf_import=pd.read_csv('./Airplane_Cleaned.csv')
safety_import=pd.read_csv('./Airplane_Crashes_and_Fatalities_Since_1908_20190820105639.csv')
price_import=pd.read_csv('./aircraft_data.csv')
safety_import
```

```python
safety_import.rename(columns={"AC Type": "Model"},inplace=True)
safety_import.info()
```

```python
prf_import
```

```python
price_import_filtered = price_import[price_import['Currency'] == 'USD']
price_import_filtered
```

```python
prf_import.columns = prf_import.columns.str.replace(' ','_')
safety_import.columns = safety_import.columns.str.replace(' ','_')
price_import_filtered.columns = price_import_filtered.columns.str.replace(' ','_')
price_import_filtered
```

```python
#Combine data into mongodb
client = pymongo.MongoClient("mongodb://localhost:27017")
```

```python
#set up tables in db
db = client.air
collection = db.price
price_import_filtered.reset_index(inplace=True)
data_dict = price_import_filtered.to_dict('records')
collection.insert_many(data_dict)
```

```python
db = client.air
collection = db.safety
safety_import.reset_index(inplace=True)
data_dict = safety_import.to_dict('records')
collection.insert_many(data_dict)
```

```python
db = client.air
collection = db.performance
prf_import.reset_index(inplace=True)
data_dict = prf_import.to_dict('records')
collection.insert_many(data_dict)
```
